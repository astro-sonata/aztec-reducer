```python
import os, glob
import matplotlib.pyplot as plt
from astropy.io import fits

```

```python
redpath = os.path.join(os.path.dirname(os.path.dirname(os.getcwd())), 'bok', '2023Dec14', 'reduced')
redfiles = glob.glob(os.path.join(redpath, '*.fits'))
print(redpath)
for file in redfiles:
    with fits.open(file) as f:
        wave, flux = f[0].data # unpack the 0th HDU

    fig, ax = plt.subplots()
    ax.plot(wave, flux)
    ax.set_xlabel(r'Wavelength [$\AA$]')
    ax.set_ylabel(r'Flux [ers s${-1}$ cm^{-2}]')
    ax.set_title(os.path.basename(file).split('.')[0].split())
```
